```python
fig,axes = plt.subplots(2,2,figsize=(10,8))
sns.boxplot(x= df['Species'],y=df['SepalLengthCm'],ax= axes[0,0])
axes[0,0].set_title("Sepal Length by Species")

sns.boxplot(x= df['Species'],y=df['SepalWidthCm'],ax= axes[0,1])
axes[0,1].set_title("Sepal Width by Species")

sns.boxplot(x= df['Species'],y=df['PetalLengthCm'],ax= axes[1,0])
axes[1,0].set_title("Petal Length by Species")

sns.boxplot(x= df['Species'],y = df['PetalWidthCm'],ax= axes[1,1])
axes[1,1].set_title("Petal Width by Species")

plt.tight_layout()
plt.show()
```

```python
sns.color_palette("Set2",as_cmap= True)


fig,axes = plt.subplots(2,2,figsize=(10,8))
sns.boxplot(x= df['Species'],y=df['SepalLengthCm'],ax= axes[0,0])
axes[0,0].set_title("Sepal Length by Species")

sns.boxplot(x= df['Species'],y=df['SepalWidthCm'],ax= axes[0,1])
axes[0,1].set_title("Sepal Width by Species")

sns.boxplot(x= df['Species'],y=df['PetalLengthCm'],ax= axes[1,0])
axes[1,0].set_title("Petal Length by Species")

sns.boxplot(x= df['Species'],y = df['PetalWidthCm'],ax= axes[1,1])
axes[1,1].set_title("Petal Width by Species")

plt.tight_layout()
plt.show()
```

```python
sns.set_palette("cubehelix")  # Use Set2 or any other palette you like

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Boxplots with hue based on species — Seaborn will use the active palette
sns.boxplot(x='Species', y='SepalLengthCm', data=df, ax=axes[0, 0])
axes[0, 0].set_title("Sepal Length by Species")

sns.boxplot(x='Species', y='SepalWidthCm', data=df, ax=axes[0, 1])
axes[0, 1].set_title("Sepal Width by Species")

sns.boxplot(x='Species', y='PetalLengthCm', data=df, ax=axes[1, 0])
axes[1, 0].set_title("Petal Length by Species")

sns.boxplot(x='Species', y='PetalWidthCm', data=df, ax=axes[1, 1])
axes[1, 1].set_title("Petal Width by Species")

# Adjust layout
plt.tight_layout()
```
